```python
# Bibliotecas de manipulação de dados
import pandas as pd
import numpy as np

# Bibliotecas de visualização
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import seaborn as sns

# Nuvem de palavras
from wordcloud import WordCloud, STOPWORDS

# Utilitários de texto
import re
from collections import Counter

# Configurações gerais
import warnings
warnings.filterwarnings('ignore')

# Configurações visuais
sns.set_theme(style='whitegrid', palette='muted')
plt.rcParams['figure.dpi'] = 110
pd.set_option('display.max_colwidth', 100)

```

```python
# Funções de análise de reviews
def reviews_histogram(reviews_in, maxcount= 100000) :
    
    # Contagem de reviews por usuário
    counts = reviews_in['reviewers'].value_counts()
    
    # Filtra apenas quem tem ≤ maxcount reviews
    filtered = counts[counts <= maxcount]
    
    # Histograma
    plt.figure(figsize=(4, 2))
    plt.hist(filtered, bins=range(0, int(maxcount *1.1), 5), color='teal', edgecolor='black')
    plt.yscale('log')  # ← Escala logarítmica no eixo y
    plt.title('Usuários com até '+str(maxcount)+' reviews')
    plt.xlabel('Número de reviews')
    plt.ylabel('Quantidade de usuários')
    plt.xticks(range(0, int(maxcount *1.1), 10))
    plt.grid(True, alpha=0.3)
    plt.show()

def reviews_quantis(reviews_in):
    # Contagem de reviews por usuário
    counts = reviews_in['reviewers'].value_counts()
    
    # Quintis (0%, 25%, 50%, 75%, 100%) + alguns extras úteis
    quintis = counts.quantile([0, 0.25, 0.5, 0.75, 0.9, 0.95, 0.99,0.995,0.999,0.9995, 1.0])
    
    print("Quantis do número de reviews por usuário:")
    print(quintis.round(0).astype(int)) 

def reviews_top(reviews_in):
    top10 = reviews_in['reviewers'].value_counts().head(5)

    plt.figure(figsize=(4, 2))
    plt.barh(top10.index, top10.values, color='skyblue')
    plt.xlabel('Quantidade de avaliações')
    plt.title('Top 5 Reviewers')
    plt.gca().invert_yaxis()  # maior no topo
    for i, v in enumerate(top10.values):
        plt.text(v + 5, i, str(v), va='center')
    plt.show()
```

## 4. Pipeline bifurcado — Merge com catálogo de cursos

> **Ramo A (dedup/qtl ~493K)** — identidade do usuário confiável → CF (SVD, KNN)
> **Ramo B (sanitizada ~1.45M)** — volume textual máximo → Content-based (TF-IDF)

```python
# --- RAMO A: merge dedup/qtl + cursos ---
reviews_A_df = reviews_dedup_qtl_df.merge(courses_df, on='course_id', how='left')

# Verifica órfãos (reviews sem curso correspondente)
orfaos_A = reviews_A_df[reviews_A_df['name'].isna()]
print(f"Ramo A — merge concluído: {len(reviews_A_df):,} reviews")
print(f"Órfãos (reviews sem curso): {len(orfaos_A):,}")

# Remove órfãos se existirem
if len(orfaos_A) > 0:
    print(f"  course_ids órfãos: {orfaos_A['course_id'].unique()}")
    reviews_A_df = reviews_A_df.dropna(subset=['name'])
    print(f"  Após remoção: {len(reviews_A_df):,} reviews")

reviews_A_df.to_parquet('../data/reviews_A_merged.parquet')
display(reviews_A_df.head(3))
print(f"\nUsuários únicos: {reviews_A_df['reviewers'].nunique():,}")
print(f"Cursos únicos:   {reviews_A_df['course_id'].nunique():,}")
```

```python
# --- RAMO B: merge sanitizada completa + cursos ---
reviews_B_df = reviews_san_df.merge(courses_df, on='course_id', how='left')

# Verifica órfãos
orfaos_B = reviews_B_df[reviews_B_df['name'].isna()]
print(f"Ramo B — merge concluído: {len(reviews_B_df):,} reviews")
print(f"Órfãos (reviews sem curso): {len(orfaos_B):,}")

if len(orfaos_B) > 0:
    print(f"  course_ids órfãos: {orfaos_B['course_id'].unique()}")
    reviews_B_df = reviews_B_df.dropna(subset=['name'])
    print(f"  Após remoção: {len(reviews_B_df):,} reviews")

# Filtra reviews de texto curto (<5 palavras) — não acrescentam sinal ao TF-IDF
reviews_B_df['word_count'] = reviews_B_df['reviews'].fillna('').str.split().str.len()
reviews_B_text_df = reviews_B_df[reviews_B_df['word_count'] >= 5].copy()

print(f"\nRamo B após filtro texto curto (<5 palavras):")
print(f"  Antes: {len(reviews_B_df):,} | Após: {len(reviews_B_text_df):,}")
print(f"  Removidos: {len(reviews_B_df) - len(reviews_B_text_df):,} ({(len(reviews_B_df) - len(reviews_B_text_df))/len(reviews_B_df)*100:.1f}%)")

reviews_B_df.to_parquet('../data/reviews_B_merged.parquet')
display(reviews_B_text_df.head(3))
```

## 5. Filtragem de esparsidade — USS / ISS (Ramo A)

Seguindo Suryani et al. (2024): removemos usuários e cursos com esparsidade acima do limiar.
- **USS** (User Specific Sparsity): `1 - (n_u / max(n_u))`
- **ISS** (Item Specific Sparsity): `1 - (n_i / max(n_i))`
- Limiar: ≤ 0.96 (manter quem avaliou ≥ 4% do máximo)

```python
SPARSITY_THRESHOLD = 0.96

# --- USS: User Specific Sparsity ---
user_counts = reviews_A_df.groupby('reviewers').size()
max_user = user_counts.max()
uss = 1 - (user_counts / max_user)

# Usuários que passam no filtro (esparsidade ≤ limiar)
users_keep = uss[uss <= SPARSITY_THRESHOLD].index
print(f"USS — máx reviews por usuário: {max_user}")
print(f"  Mín reviews para passar (limiar {SPARSITY_THRESHOLD}): {int(max_user * (1 - SPARSITY_THRESHOLD))}")
print(f"  Usuários antes: {len(uss):,} | Após: {len(users_keep):,}")
print(f"  Removidos: {len(uss) - len(users_keep):,}")

# --- ISS: Item Specific Sparsity ---
item_counts = reviews_A_df.groupby('course_id').size()
max_item = item_counts.max()
iss = 1 - (item_counts / max_item)

items_keep = iss[iss <= SPARSITY_THRESHOLD].index
print(f"\nISS — máx reviews por curso: {max_item}")
print(f"  Mín reviews para passar (limiar {SPARSITY_THRESHOLD}): {int(max_item * (1 - SPARSITY_THRESHOLD))}")
print(f"  Cursos antes: {len(iss):,} | Após: {len(items_keep):,}")
print(f"  Removidos: {len(iss) - len(items_keep):,}")

# Aplica ambos os filtros
reviews_A_filt_df = reviews_A_df[
    reviews_A_df['reviewers'].isin(users_keep) &
    reviews_A_df['course_id'].isin(items_keep)
].copy()

print(f"\nResultado final do filtro USS/ISS:")
print(f"  Reviews: {len(reviews_A_df):,} → {len(reviews_A_filt_df):,}")
print(f"  Usuários: {reviews_A_filt_df['reviewers'].nunique():,}")
print(f"  Cursos:   {reviews_A_filt_df['course_id'].nunique():,}")
print(f"  (Ref. Suryani: ~233 cursos, ~39K registros)")

reviews_A_filt_df.to_parquet('../data/reviews_A_filtered.parquet')
```

```python
# Distribuição de USS e ISS para visualização
fig, axes = plt.subplots(1, 2, figsize=(10, 3))

axes[0].hist(uss, bins=50, color='teal', edgecolor='black', alpha=0.7)
axes[0].axvline(SPARSITY_THRESHOLD, color='red', linestyle='--', label=f'Limiar {SPARSITY_THRESHOLD}')
axes[0].set_title('Distribuição USS (por usuário)')
axes[0].set_xlabel('Esparsidade')
axes[0].set_ylabel('Contagem')
axes[0].legend()

axes[1].hist(iss, bins=50, color='coral', edgecolor='black', alpha=0.7)
axes[1].axvline(SPARSITY_THRESHOLD, color='red', linestyle='--', label=f'Limiar {SPARSITY_THRESHOLD}')
axes[1].set_title('Distribuição ISS (por curso)')
axes[1].set_xlabel('Esparsidade')
axes[1].set_ylabel('Contagem')
axes[1].legend()

plt.tight_layout()
plt.show()
```

## 6. Matriz pivô, formato esparso e normalização (Ramo A)

Construímos a matriz usuário × curso com ratings. Convertemos para CSR (Compressed Sparse Row) para eficiência de memória. A versão normalizada (mean-centering) corrige o viés positivo dos ratings.

```python
from scipy import sparse

# Matriz pivô: usuários nas linhas, cursos nas colunas, ratings nos valores
pivot_df = reviews_A_filt_df.pivot_table(
    index='reviewers',
    columns='course_id',
    values='rating',
    aggfunc='last'  # último rating caso reste duplicata residual
)

print(f"Matriz pivô: {pivot_df.shape[0]:,} usuários × {pivot_df.shape[1]:,} cursos")

# Densidade real
total_cells = pivot_df.shape[0] * pivot_df.shape[1]
filled_cells = pivot_df.notna().sum().sum()
density = filled_cells / total_cells
print(f"Células preenchidas: {filled_cells:,} / {total_cells:,}")
print(f"Densidade: {density:.4f} ({density*100:.2f}%)")
print(f"Esparsidade: {1 - density:.4f} ({(1 - density)*100:.2f}%)")

# Converte para CSR (NaN → 0 no formato esparso)
pivot_filled = pivot_df.fillna(0)
matrix_csr = sparse.csr_matrix(pivot_filled.values)
print(f"\nCSR: {matrix_csr.shape}, nnz={matrix_csr.nnz:,}, dtype={matrix_csr.dtype}")

# Versão normalizada (mean-centering por usuário)
user_means = pivot_df.mean(axis=1)  # média por usuário (ignora NaN)
pivot_norm_df = pivot_df.sub(user_means, axis=0)
pivot_norm_filled = pivot_norm_df.fillna(0)
matrix_norm_csr = sparse.csr_matrix(pivot_norm_filled.values)

print(f"\nMean-centering aplicado:")
print(f"  Média das médias por usuário: {user_means.mean():.2f}")
print(f"  Rating médio normalizado: {pivot_norm_df.stack().mean():.4f} (esperado ≈ 0)")

# Salva artefatos: parquet para DataFrames, npz para matrizes esparsas
pivot_df.to_parquet('../data/matrix_pivot.parquet')
sparse.save_npz('../data/matrix_csr.npz', matrix_csr)
sparse.save_npz('../data/matrix_norm_csr.npz', matrix_norm_csr)
user_means.to_frame('mean_rating').to_parquet('../data/matrix_user_means.parquet')

print("\nArtefatos salvos: matrix_pivot.parquet, matrix_csr.npz, matrix_norm_csr.npz, matrix_user_means.parquet")
```

## 7. Divisão temporal treino / teste (Ramo A)

Split conforme EDA: treino = reviews antes de 2020, teste = 2020 em diante. O pico pandêmico simula cenário real de crescimento de novos usuários.

```python
SPLIT_DATE = pd.Timestamp('2020-01-01')

train_df = reviews_A_filt_df[reviews_A_filt_df['date_reviews'] < SPLIT_DATE].copy()
test_df  = reviews_A_filt_df[reviews_A_filt_df['date_reviews'] >= SPLIT_DATE].copy()

print(f"Split temporal em {SPLIT_DATE.date()}:")
print(f"  Treino: {len(train_df):,} reviews ({len(train_df)/len(reviews_A_filt_df)*100:.1f}%)")
print(f"  Teste:  {len(test_df):,} reviews ({len(test_df)/len(reviews_A_filt_df)*100:.1f}%)")

# Verifica órfãos: usuários/cursos presentes só no teste
train_users  = set(train_df['reviewers'].unique())
test_users   = set(test_df['reviewers'].unique())
train_courses = set(train_df['course_id'].unique())
test_courses  = set(test_df['course_id'].unique())

orphan_users   = test_users - train_users
orphan_courses = test_courses - train_courses

print(f"\nÓrfãos (presentes só no teste):")
print(f"  Usuários: {len(orphan_users):,} (cold start potencial)")
print(f"  Cursos:   {len(orphan_courses):,}")

# Mantemos órfãos no teste — são cenários reais de cold start
# que o componente content-based (Ramo B) deve cobrir
print(f"\nTreino — usuários: {len(train_users):,} | cursos: {len(train_courses):,}")
print(f"Teste  — usuários: {len(test_users):,} | cursos: {len(test_courses):,}")

train_df.to_parquet('../data/train_reviews.parquet')
test_df.to_parquet('../data/test_reviews.parquet')
```

```python
# Matrizes pivô separadas para treino e teste

# Treino
train_pivot = train_df.pivot_table(
    index='reviewers', columns='course_id', values='rating', aggfunc='last'
)
train_means = train_pivot.mean(axis=1)
train_norm = train_pivot.sub(train_means, axis=0).fillna(0)
train_csr = sparse.csr_matrix(train_pivot.fillna(0).values)
train_norm_csr = sparse.csr_matrix(train_norm.values)

# Teste
test_pivot = test_df.pivot_table(
    index='reviewers', columns='course_id', values='rating', aggfunc='last'
)

print(f"Matriz treino: {train_pivot.shape[0]:,} × {train_pivot.shape[1]:,}")
print(f"Matriz teste:  {test_pivot.shape[0]:,} × {test_pivot.shape[1]:,}")

# Salva artefatos de treino (parquet + npz)
train_pivot.to_parquet('../data/train_pivot.parquet')
sparse.save_npz('../data/train_csr.npz', train_csr)
sparse.save_npz('../data/train_norm_csr.npz', train_norm_csr)
train_means.to_frame('mean_rating').to_parquet('../data/train_user_means.parquet')

# Salva artefatos de teste
test_pivot.to_parquet('../data/test_pivot.parquet')

print("\nArtefatos de treino/teste salvos (parquet + npz)")
```

## 8. TF-IDF para content-based / cold start (Ramo B)

Usamos a base sanitizada completa (~1.45M) para maximizar o volume textual. Para cada curso, concatenamos: nome + instituição + texto agregado dos reviews (≥5 palavras). O TF-IDF resultante alimenta o componente content-based do sistema híbrido.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import json

# Agrega texto dos reviews (≥5 palavras) por curso
text_by_course = (
    reviews_B_text_df
    .groupby('course_id')['reviews']
    .apply(lambda x: ' '.join(x.dropna()))
    .reset_index()
    .rename(columns={'reviews': 'review_text'})
)

# Junta metadados do catálogo
course_corpus = courses_df[['course_id', 'name', 'institution']].merge(
    text_by_course, on='course_id', how='left'
)

# Corpus final: nome + instituição + reviews agregados
course_corpus['corpus'] = (
    course_corpus['name'].fillna('') + ' ' +
    course_corpus['institution'].fillna('') + ' ' +
    course_corpus['review_text'].fillna('')
)

print(f"Cursos com texto de reviews: {course_corpus['review_text'].notna().sum()} / {len(course_corpus)}")
print(f"Cursos sem reviews (só metadados): {course_corpus['review_text'].isna().sum()}")

# TF-IDF
tfidf = TfidfVectorizer(
    max_features=5000,
    stop_words='english',
    ngram_range=(1, 2),
    min_df=2
)
tfidf_matrix = tfidf.fit_transform(course_corpus['corpus'])

print(f"\nMatriz TF-IDF: {tfidf_matrix.shape[0]} cursos × {tfidf_matrix.shape[1]} features")
print(f"Densidade: {tfidf_matrix.nnz / (tfidf_matrix.shape[0] * tfidf_matrix.shape[1]):.4f}")

# Top termos por peso médio
feature_names = tfidf.get_feature_names_out()
mean_tfidf = tfidf_matrix.mean(axis=0).A1
top_idx = mean_tfidf.argsort()[-15:][::-1]
print(f"\nTop 15 termos por peso médio TF-IDF:")
for i in top_idx:
    print(f"  {feature_names[i]:25s} {mean_tfidf[i]:.4f}")

# Salva artefatos TF-IDF: npz para matriz esparsa, parquet para índice, JSON para vocabulário
sparse.save_npz('../data/tfidf_matrix.npz', tfidf_matrix)

tfidf_meta = pd.DataFrame({'course_id': course_corpus['course_id'].tolist()})
tfidf_meta.to_parquet('../data/tfidf_meta.parquet', index=False)

# Salva parâmetros do vectorizer como JSON (evita pickle — compatível entre versões numpy)
tfidf_params = {
    'vocabulary': {k: int(v) for k, v in tfidf.vocabulary_.items()},
    'idf': tfidf.idf_.tolist(),
    'max_features': 5000,
    'stop_words': 'english',
    'ngram_range': [1, 2],
    'min_df': 2
}
with open('../data/tfidf_params.json', 'w') as f:
    json.dump(tfidf_params, f)

print("\ntfidf_matrix.npz, tfidf_meta.parquet, tfidf_params.json salvos")
```

## 9. Preparação para Surprise — triplas (user, item, rating) (Ramo A)

A biblioteca Surprise espera dados no formato de triplas. Criamos datasets de treino e teste prontos para alimentar KNN e SVD.

```python
# Triplas para Surprise: (reviewers, course_id, rating)
surprise_train = train_df[['reviewers', 'course_id', 'rating']].copy()
surprise_test  = test_df[['reviewers', 'course_id', 'rating']].copy()

# Garante tipos corretos
surprise_train['course_id'] = surprise_train['course_id'].astype(str)
surprise_test['course_id']  = surprise_test['course_id'].astype(str)

print(f"Surprise treino: {len(surprise_train):,} triplas")
print(f"Surprise teste:  {len(surprise_test):,} triplas")
print(f"\nRating scale: {surprise_train['rating'].min()} – {surprise_train['rating'].max()}")

# Salva como parquet e CSV (para reprodutibilidade)
surprise_train.to_parquet('../data/surprise_train.parquet')
surprise_test.to_parquet('../data/surprise_test.parquet')
surprise_train.to_csv('../data/surprise_train.csv', index=False)
surprise_test.to_csv('../data/surprise_test.csv', index=False)

print("\nsurprise_train.parquet/csv e surprise_test.parquet/csv salvos")
display(surprise_train.head())
```

## 10. Resumo estatístico final do pipeline

```python
# Resumo comparativo de todas as etapas do pipeline
print("=" * 65)
print("RESUMO DO PIPELINE")
print("=" * 65)

stages = [
    ("Reviews brutos",        len(reviews_df)),
    ("Sanitizados",           len(reviews_san_df)),
    ("Dedup (Ramo A)",        len(reviews_dedup_df)),
    ("Dedup+Qtl (Ramo A)",    len(reviews_dedup_qtl_df)),
    ("Merged (Ramo A)",       len(reviews_A_df)),
    ("Filtrado USS/ISS",      len(reviews_A_filt_df)),
    ("  ├─ Treino (<2020)",   len(train_df)),
    ("  └─ Teste  (≥2020)",   len(test_df)),
]

print(f"\n{'Etapa':<30s} {'Reviews':>10s}")
print("-" * 42)
for name, count in stages:
    print(f"{name:<30s} {count:>10,}")

print(f"\n{'Ramo B (sanitizada completa)':<30s} {len(reviews_B_df):>10,}")
print(f"{'  └─ c/ texto ≥5 palavras':<30s} {len(reviews_B_text_df):>10,}")

print(f"\n{'--- DADOS FINAIS (Ramo A) ---':^42}")
print(f"Usuários (treino): {train_df['reviewers'].nunique():,}")
print(f"Cursos (treino):   {train_df['course_id'].nunique():,}")
print(f"Matriz treino:     {train_pivot.shape}")
train_density = train_pivot.notna().sum().sum() / (train_pivot.shape[0] * train_pivot.shape[1])
print(f"Densidade treino:  {train_density:.4f} ({train_density*100:.2f}%)")

print(f"\n{'--- DISTRIBUIÇÃO DE RATINGS ---':^42}")
for split_name, split_data in [("Treino", train_df), ("Teste", test_df)]:
    dist = split_data['rating'].value_counts().sort_index()
    total = len(split_data)
    print(f"\n{split_name}:")
    for rating, count in dist.items():
        bar = "█" * int(count / total * 40)
        print(f"  {rating}★: {count:>6,} ({count/total*100:5.1f}%) {bar}")

print(f"\n{'--- TF-IDF (Ramo B) ---':^42}")
print(f"Cursos no corpus:  {tfidf_matrix.shape[0]}")
print(f"Features TF-IDF:   {tfidf_matrix.shape[1]}")

print(f"\n{'--- ARTEFATOS GERADOS ---':^42}")
import os
data_dir = '../data/'
for fname in sorted(os.listdir(data_dir)):
    if fname.endswith(('.parquet', '.npz', '.json', '.csv')):
        size = os.path.getsize(os.path.join(data_dir, fname))
        if size > 1024*1024:
            print(f"  {fname:<35s} {size/1024/1024:.1f} MB")
        else:
            print(f"  {fname:<35s} {size/1024:.0f} KB")
```
